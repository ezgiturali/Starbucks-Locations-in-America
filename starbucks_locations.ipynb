```python
import bokeh
```

```python
from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure
from bokeh.models import GeoJSONDataSource, LinearColorMapper, ColorBar
from bokeh.palettes import brewer
import json
```

```python
merged
```

```python
merged_json = json.loads(merged.to_json())
json_data = json.dumps(merged_json)
```

```python
geosource = GeoJSONDataSource(geojson = json_data)
```

```python
palette = bokeh.palettes.YlOrRd9
```

```python
color_mapper = LinearColorMapper(palette = palette,   low=0, high=3000,nan_color = '#d9d9d9')
```

```python
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8,width = 500, height = 20,
border_line_color='black',location = (0,0), orientation ='horizontal')
```

```python
p = figure(title = 'Starbucks', plot_height = 700 , plot_width = 700, toolbar_location = None, 
          tooltips=[
         ("State", "@name"),
         ("Number of Starbucks", "@NumberofBucks")])
```

```python
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
```

```python
p.patches('xs','ys', source = geosource,fill_color = {'field':'NumberofBucks', 'transform':color_mapper},
         line_color = 'black', line_width = 0.25, fill_alpha = 1)
```

```python
p.add_layout(color_bar, 'below')
```

```python
output_notebook()
```

```python
show(p)
```
